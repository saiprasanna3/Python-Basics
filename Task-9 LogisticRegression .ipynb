```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn 
```

```python
df=pd.read_csv('C:\\Users\\saipr\\Downloads\\HR_comma_sep.csv')
```

```python
#printing dataset
df.head()
```

```python
#Logistic Regression model
#takin
df1=df[['salary','satisfaction_level','average_montly_hours','promotion_last_5years','left']]
```

```python
#getting dummies
dummies=pd.get_dummies(df.salary)
```

```python
dummies
```

```python
#adding dummies column to df1 to get numerical data
df1=pd.concat([df1,dummies],axis='columns')
```

```python
df1.head()
```

```python
#dummy variable trap
df1=df1.drop(['salary','medium'],axis=1)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df1[['satisfaction_level','average_montly_hours','promotion_last_5years','high','low']],df1.left,test_size=0.75,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

# Accuracy

```python
#predicting score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
