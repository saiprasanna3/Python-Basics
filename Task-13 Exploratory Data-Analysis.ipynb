```python
df['Age']=df['Age'].fillna(method='ffill')
```

```python
#Filling missing values of Embarked column
df['Embarked'].unique()
```

```python
df['Embarked']=df['Embarked'].replace(np.nan,'S')
```

```python
df['Embarked'].isnull().sum()
```

```python
df['Embarked'].unique()
```

```python
#as there are many null values in Cabin column it is better to drop it
df=df.drop('Cabin',axis='columns')
```

```python
df
```

```python
df.isnull().sum()
```

```python
#Creating new column by adding SibSp and Parch
df["FamilySize"] = df["SibSp"] + df["Parch"]
df.head()
```

```python
#Dropping the columns of SibSp and Parch
df=df.drop(['SibSp','Parch'],axis=1)
```
