```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df
```

```python
df['target']=digits.target
df
```

```python
x=df.drop(['target'],axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.75,random_state=1)
```

```python
print(len(x_train))
print(len(x_test))
```

# Checking score of rbf kernel

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf',random_state=0,probability=True)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print('Model Score of rbf:',model.score(x_test,y_test))
```

# Checking score of linear kernel

```python
from sklearn.svm import SVC
model=SVC(kernel='linear',random_state=0,probability=True)
model.fit(x_train,y_train)
y_pred1=model.predict(x_test)
print('Model Score of linear:',model.score(x_test,y_test))
```

# Checking score of poly kernel

```python
from sklearn.svm import SVC
model=SVC(kernel='poly',random_state=0,probability=True)
model.fit(x_train,y_train)
y_pred2=model.predict(x_test)
print('Model Score of poly:',model.score(x_test,y_test))
```

```python
#poly kernel gave best score ,let's check its confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix
cm(y_test,y_pred2)
```
